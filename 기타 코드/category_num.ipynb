```python
import pandas as pd
import requests as req  
import json

from bs4 import BeautifulSoup as bfs
from tqdm.notebook import tqdm
```

```python
def num_print():
  url = 'https://store.musinsa.com/app/svc/get_category_list_jsonp?cat_type=ITEM&sex=&callback=__jp0'
  headers = {
      'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.57 Whale/3.14.133.23 Safari/537.36',
      'referer': 'https://www.musinsa.com/'
  }

  resp = req.get(url=url, headers=headers)

  dic = json.loads(resp.text[6:-2])
  h_num=list(dic['category'].keys())

  h_cate = {}
  l_cate = {}
  for h in h_num:
    h_cate[dic['category'][h]['title']] = h

    l_num = list(dic['category'][h]['childs'].keys())
    for l in l_num:
      l_cate[dic['category'][h]['childs'][l]['title']] = l

  a = input('상위, 하위 입력: ')
  if a == '상위':
    b = input('카테고리 입력: ')
    if b in h_cate.keys():
      print('상품코드: {}'.format(h_cate[b]))
    else:
      print('상위 카테고리를 입력하세요.')
  elif a == '하위':
    b = input('카테고리 입력: ')
    if b in l_cate.keys():
      print('상품코드: {}'.format(l_cate[b]))
    else:
      print('하위 카테고리를 입력하세요.')
  else:
    print('상위, 하위를 입력하세요.')
```

```python
num_print()
```
